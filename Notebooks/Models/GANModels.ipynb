```python
from dpm.models import GAN, WGAN, LSGAN, MMGAN
import numpy as np
import matplotlib.pyplot as plt
from dpm.visualize import plot_stats, plot_hist, plot_model, plot_hists
from dpm.distributions import Normal
from dpm.mixture_models import MixtureModel
import torch
```

```python
X = MixtureModel([Normal(-4., 2.3, learnable=False), Normal(4., 2.3, learnable=False)], [0.5, 0.5]).sample(10000)
```

```python
plot_hist(X)
```

```python
X = X.numpy()
```

```python
model = GAN({'input_dim':8}, {'input_dim':1, 'use_spectral_norm':True})
stats = model.fit(X, epochs=10000, lr=1e-4)
plot_stats(stats)
```

```python
preds = model.sample(10000)
plot_hist(preds.detach())
```

```python
model = WGAN({'input_dim':8}, {'input_dim':1, 'use_spectral_norm':True})
stats = model.fit(X, epochs=10000, lr=1e-4)
plot_stats(stats)
```

```python
preds = model.sample(10000)
plot_hist(preds.detach())
```

```python
model = MMGAN({'input_dim':8}, {'input_dim':1, 'use_spectral_norm':True})
stats = model.fit(X, epochs=10000, lr=1e-4)
plot_stats(stats)
```

```python
preds = model.sample(10000)
plot_hist(preds.detach())
```
