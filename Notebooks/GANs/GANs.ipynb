```python
from dpm.train import train
from dpm.distributions import Generator, Data, Normal
from dpm.criterion import GANLoss, WGANLoss, MMGANLoss, LSGANLoss
from dpm.distributions import MixtureModel
from dpm.visualize import plot_model, plot_models, plot_stats
import matplotlib.pyplot as plt
```

```python
p_model = Data(MixtureModel([Normal(-4., 2.3), Normal(4., 2.3)], [0.5, 0.5]).sample(10000))
q_model = Generator()

plot_models(p_model, q_model); plt.show()
```

```python
p_model = Data(MixtureModel([Normal(-4., 2.3), Normal(4., 2.3)], [0.5, 0.5]).sample(10000))
q_model = Generator()

criterion = GANLoss(1)

stats = train(p_model, q_model, criterion, epochs=20000, optimizer='RMSprop', lr=1e-4, track_parameters=False)
plot_stats(stats); plt.show()
plot_models(p_model, q_model); plt.show()
```

```python
p_model = Data(MixtureModel([Normal(-4., 2.3), Normal(4., 2.3)], [0.5, 0.5]).sample(10000))
q_model = Generator()

criterion = GANLoss(1, use_spectral_norm=True)

stats = train(p_model, q_model, criterion, epochs=20000, optimizer='RMSprop', 
              lr=1e-4, batch_size=512, track_parameters=False)
plot_stats(stats); plt.show()
plot_models(p_model, q_model); plt.show()
```

```python
p_model = Data(MixtureModel([Normal(-4., 2.3), Normal(4., 2.3)], [0.5, 0.5]).sample(10000))
q_model = Generator()

criterion = WGANLoss(1, use_spectral_norm=True)

stats = train(p_model, q_model, criterion, epochs=20000, optimizer='RMSprop', lr=1e-4, 
              batch_size=512,
              track_parameters=False)
plot_stats(stats); plt.show()
plot_models(p_model, q_model); plt.show()
```

```python
p_model = Data(MixtureModel([Normal(-4., 2.3), Normal(4., 2.3)], [0.5, 0.5]).sample(10000))
q_model = Generator()

criterion = WGANLoss(1, grad_penalty=30., use_spectral_norm=False)

stats = train(p_model, q_model, criterion, epochs=20000, optimizer='RMSprop', 
              lr=1e-4, batch_size=512, track_parameters=False)
plot_stats(stats); plt.show()
plot_models(p_model, q_model); plt.show()
```
